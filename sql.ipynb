```python
#Import libraries
import numpy as np 
import pandas as pd
```

```python
# Taking a quick look at the data to see columns
trail_data = pd.read_csv("trail_Data.csv")
trail_data
```

```python
# Convert meters in length to miles
#Conver meters in elevation_gain to feet
trail_data["length"] = trail_data["length"] / 1609
trail_data["elevation_gain"] = trail_data["elevation_gain"] * 3.281

```

```python
#Quickly look at the column names
trail_data.columns.to_list()
```

```python
trail_data_SQL = trail_data
```

```python
#import sqlalchemy and create engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

#export the dataframe as a table to the sqlite engine
trail_data_SQL.to_sql("TrailData", con=engine)
```

```python
#Quick test query to ensure columns are  properly showing up
sql='''

SELECT * FROM traildata

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
#Most popular trails in all parks
sql='''

SELECT name, area_name  
FROM traildata
ORDER BY popularity DESC

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```

```python
#Most popular trails by a certain park name
#Can edit the "Where clause to pick the specific park"
sql='''

SELECT name, area_name, length  
FROM traildata
WHERE area_name = "Zion National Park" 
ORDER BY popularity DESC
;

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```

```python
#Which National Park has the highest total miles of trails
sql='''

SELECT area_name,  SUM(length) as total_miles
FROM traildata
GROUP BY area_name
Order BY total_miles DESC
;

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
#Which National Park has the most(least) number of trails 
sql='''

SELECT area_name,  COUNT(DISTINCT name) as total_trails
FROM traildata
GROUP BY area_name
Order BY total_trails DESC
;

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```

```python
#Which National Park has the highest average trail length 
sql='''

SELECT area_name,  AVG( length) as avg_length
FROM traildata
GROUP BY area_name
Order BY avg_length DESC
;

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```

```python
#Which National Park has the highest average difficulty 
sql='''

SELECT area_name,  AVG(difficulty_rating) as avg_diff
FROM traildata
GROUP BY area_name
Order BY avg_diff DESC
;

''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(10)
```
